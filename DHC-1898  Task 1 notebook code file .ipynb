```python

print("Shape of dataset:", df.shape)
print("\nColumns in dataset:")
print(df.columns)
print("\n the following are the first 5 rows:")
print(df.head())
```

_______________________________________________

now i am checking data overall info and statistics
by "describe()" i have obsrved central tendency and spread of numerical features. The dataset show ranges and no anomalies but it has the missing values in feature/column that are : bill_length_mm, bill_depth_mm, flipper_length_mm, body_mass_g and sex/gender

```python
print("Dataset infrmation:\n")
df.info()
print("\nStatistics:\n")
print(df.describe())
```

_______________________________________________________

now im doing data cleaning and handling mssing values
as i saw missing values in above step so now removing rows with missing values by using "dropna() (removing row that having missing value)" and number of missing val was very small

```python
print("printing missng values before cleaning:\n")
print(df.isnull().sum())

df_clean = df.dropna()

print("\nprinting again missing values after cleaning:\n")
print(df_clean.isnull().sum())

print("\nNew dataset shape:", df_clean.shape)
```

_________________________________________________


using a scatter plot to print the relationship between bill length and bill depth colored by species
Now The plot shows clear clustering and indicating that the species can be separated based on features
follwoing observations:
Adelie (blue): having Higher bill depth and shorter to medium bill length and cluster on left side

Gentoo (green): The lowest bill depth and higher spread in bill length as well as clearly separated at bottom

Chinstrap (orange): Higher bill length medium to high bill depth and sits between Adelie and Gentoo

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df_clean,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species"
)
plt.title("Bill Length vs Bill Depth")
plt.show()
```

______________________________________________________


Using histograms to obsrve the distribution of numerical features
So the plots show that feature like flipper length and body mass have wider variation, which can help in distinguishing between penguin species
Following observation:

bill_length_mm: Slight spread and different peaks for species

bill_depth_mm : more compact distribution

flipper_length_mm :Clear variation

body_mass_g: Wide range indicates size differences

```python
df_clean.hist(figsize=(10,8), bins=20)
plt.suptitle("Feature distributions")
plt.show()
```

_______________________________________________________


using the box plots to compare body mass of different penguin species
i have obsrved that gentoo penguins are heavier than adelie and ahinstrap species as well as no major outliers observed

```python
plt.figure(figsize=(8,6))

sns.boxplot(
    data=df_clean,
    x="species",
    y="body_mass_g"
)

plt.title("body mass distribution species")
plt.show()
```
